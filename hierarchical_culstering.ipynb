```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Mall_Customers.csv")
```

```python
data
```

```python
X = data.iloc[:,[2,3]].values
```

```python
X
```

```python
print(X)
```

```python
import scipy.cluster.hierarchy as sch
```

```python
from scipy.cluster import hierarchy as sch
```

```python
den = sch.dendrogram(sch.linkage(X, method= 'ward'))
plt.title("dendogram")
plt.xlabel("customers")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=4, metric = 'euclidean', linkage ='ward')
```

```python
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
X
```

```python
y_hc
```

```python
plt.scatter(X[y_hc == 0,0],X[y_hc == 0,1],s = 100, c = 'red' , label = 'c1')
plt.scatter(X[y_hc == 1,0],X[y_hc == 1,1],s = 100, c = 'blue' , label = 'c2')
plt.scatter(X[y_hc == 2,0],X[y_hc == 2,1],s = 100, c = 'yellow' , label = 'c3')
plt.scatter(X[y_hc == 3,0],X[y_hc == 3,1],s = 100, c = 'orange' , label = 'c4')
plt.title("Hierarchial_clustering")
plt.xlabel("annual_income")
plt.ylabel("Annual_spending")
plt.legend()
plt.show()
```
